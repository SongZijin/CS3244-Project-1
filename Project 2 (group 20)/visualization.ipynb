```python
# I have a data set with the following columns: Question Text, Topics, Difficulty Level, Array, String, Dynamic Programming, Math, Hash Table, Greedy, Sorting, Depth-First Search, Breadth-First Search, Binary Search.
# For the topic columns, the value is 1 if the question is related to the topic, and 0 otherwise.
# I want to create a visualization that shows the distribution of the topics (Array, String, Dynamic Programming, Math, Hash Table, Greedy, Sorting, Depth-First Search, Breadth-First Search, Binary Search) of the entire data set.
# First import the necessary libraries and read the data set named "trimed_leetcode_questions.csv".
# Then, create a new data frame that contains only the topic columns.
# Then, create a new data frame that contains the sum of each topic column from df.
# Then, create a visualization that shows the distribution of the topics of the entire data set.
# The visualization should be a bar chart with the topic names on the x-axis and the number of questions on the y-axis.
# The visualization should be sorted by the number of questions in descending order.
# The visualization should have a title and x-axis label and y-axis label.
# The visualization should be saved as "topic_distribution.png".
# The visualization should be saved in the same folder as this notebook.
# The visualization should be 10 inches wide and 5 inches tall.
# The visualization should have a white background.
# The visualization should have a grid.
# The visualization should have a legend.
# Make sure the labels are readable and do not overlap.
# code
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('trimed_leetcode_questions.csv')
df_topic = df[['Array', 'String', 'Dynamic Programming', 'Math', 'Hash Table', 'Greedy', 'Sorting', 'Depth-First Search', 'Breadth-First Search', 'Binary Search']]
df_topic_sum = df_topic.sum()
df_topic_sum = df_topic_sum.sort_values(ascending=False)
plt.figure(figsize=(10,5))
plt.bar(df_topic_sum.index, df_topic_sum.values)
plt.title('Topic Distribution')
plt.xlabel('Topic')
plt.xticks(rotation=90)
plt.ylabel('Number of Questions')
plt.grid(axis='y')
plt.savefig('topic_distribution.png')
for i, v in enumerate(df_topic_sum.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

```
